```python
plt.scatter(df['Sale'], df['Price'])
z = np.polyfit(df['Sale'], df['Price'], 1)
p = np.poly1d(z)
plt.plot(df['Sale'], p(df['Sale']), 'r--')
plt.xlabel('Sale')
plt.ylabel('Price')
plt.title('Relationship between Sales and Prices')
plt.show()
```
